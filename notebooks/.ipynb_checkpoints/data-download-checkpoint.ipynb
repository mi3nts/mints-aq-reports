```julia
using CSV, DataFrames 
using Plots
using Markdown
```

```julia
df_BME280 = CSV.File(download("https://ncsa.osn.xsede.org/ees230012-bucket01/central-node-8/2023/2023/05/02/MINTS_001e06323a37_BME280_2023_05_02.csv")) |> DataFrame
```

```julia
df_IPS7100 = CSV.File(download("https://ncsa.osn.xsede.org/ees230012-bucket01/central-node-8/2023/2023/05/02/MINTS_001e06323a37_IPS7100_2023_05_02.csv")) |> DataFrame
```

```julia
#| echo: false

md"""

This is a sample markdown string 

\begin{equation}
    \int_a^b f(x)dx 
\end{equation}

We can do equations!

Further, we can impute values. For example, the maximum PM 2.5 concentration for 5-2-2023 was **$(maximum(df_IPS7100.pm2_5)) μg/m³**

Let's create a simple plot: 
"""
```

```julia
#| echo: false
#| output: false

using Dates

function date2datetime(dt::AbstractString;  df = dateformat"yyyy-mm-dd HH:MM:SS.sss")
    dt_string = String(dt)
    dt_split = split(dt_string, ".")
    dt_millisecond = join([dt_split[1], dt_split[2][1:end-3]], ".")
    return DateTime(dt_millisecond, df)
end

date2datetime(df_IPS7100.dateTime[1])
```

```julia
plot(
    date2datetime.(df_IPS7100.dateTime),
    df_IPS7100.pm2_5,
    ylabel="Concentration [μg/m³]",
    title="PM 2.5 at Central Node 8",
    lw=3,
    label="",
    xrotation=45,
    margin=10*Plots.mm,
    size=(900,650),
)
```
